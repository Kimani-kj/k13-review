```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)  # Create a line plot
plt.title('Sample Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()  # Display the plot
```

```python
plt.scatter(x, y)  # Create a scatter plot
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)  # Create a 2x1 grid of subplots
ax[0].plot(x, y)
ax[1].scatter(x, y)
plt.show()
```
