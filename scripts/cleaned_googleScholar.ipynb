```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Function to extract specific countries and exclude certain terms (case-insensitive)
def extract_countries(text, countries_list, exclude_terms=None):
    if exclude_terms is None:
        exclude_terms = []  # Default to empty list if no exclusions specified
    
    # Convert countries list and exclude terms to lowercase for case-insensitive matching
    countries_list = [country.lower() for country in countries_list]
    exclude_terms = [term.lower() for term in exclude_terms]
    
    doc = nlp(text)  # Process the text using spaCy NLP model
    countries = [
        ent.text for ent in doc.ents
        if ent.label_ == 'GPE' and ent.text.lower() in countries_list and ent.text.lower() not in exclude_terms
    ]
    return countries

# Define the list of countries to look for (case-insensitive)
countries_list = ["Algeria", "Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", "Cabo Verde", "Cameroon", "Central African Republic", "Chad", "Comoros", "Congo", "Democratic Republic of the Congo", "Djibouti", "Egypt", "Equatorial Guinea", "Eritrea", "Eswatini", "Ethiopia", "Gabon", "Gambia", "Ghana", "Guinea", "Guinea-Bissau", "Ivory Coast", "Kenya", "Lesotho", "Liberia", "Libya", "Madagascar", "Malawi", "Mali", "Mauritania", "Mauritius", "Morocco", "Mozambique", "Namibia", "Niger", "Nigeria", "Rwanda", "São Tomé and Príncipe", "Senegal", "Seychelles", "Sierra Leone", "Somalia", "South Africa", "South Sudan", "Sudan", "Togo", "Tunisia", "Uganda",'tanzania','rwanda', 'burundi' "Zambia", "Zimbabwe", "Western Sahara"]


# List of terms to exclude from the extracted countries (case-insensitive)
exclude_terms = ['K13', 'Isolates', 'Malaria', 'Artemisia']  # Add any terms you want to exclude

# Ensure 'df_noDuplicates' is defined and has the 'Title' column
if 'Title' in df.columns:
    # Apply the extract_countries function to the 'Title' column
    df['countries'] = df['Title'].apply(lambda text: extract_countries(text, countries_list, exclude_terms=exclude_terms))
    
    # Display the resulting DataFrame
    print(df[['Title', 'countries']].head())  # Show the first few rows for inspection
else:
    print("Error: 'Title' column not found in the DataFrame.")
```
