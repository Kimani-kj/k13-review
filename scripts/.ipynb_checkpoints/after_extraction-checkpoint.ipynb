```python
df=pd.read_excel('./cleaned_K13.xlsx')
df
```

```python
country=df.groupby("country")["pmid"].nunique()
country_df=pd.DataFrame(country, index=None)
plt.figure(figsize=(10,6))
plt.xticks(rotation=45)
bars=plt.bar(country_df.index, country_df["pmid"])
```

```python
# Group by 'country' and calculate the number of unique 'pmid' for each country
country = df.groupby("country")["pmid"].nunique()

# Convert the Series to a DataFrame (index is automatically set by the groupby operation)
country_df = pd.DataFrame(country)

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(country_df.index, country_df["pmid"],)

for bar in bars:
    yval = bar.get_height()  # Get the height (number of unique PMIDs) of each bar
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha="center", va="bottom")  # Label the bar
    
sns.despine(top=True, right=True)

# Adding labels and title
#plt.xlabel("Country")
plt.ylabel("Number of papers")
plt.title("paper per country")

# Show plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
df
```

```python
pinot_validate=df.pivot_table(
    index="country",
    columns="k13_validated_mutations",
    aggfunc="count"
)
pinot_validate
```

```python
pivot_table = df.pivot_table(
    index='country', 
    columns='k13_validated_mutations', 
    aggfunc='size', 
    fill_value=0
)
pivot_table
```

```python
pivot_table.drop(columns=["-",'NONE'], inplace=True)
pivot_table
```

```python
# Plotting the pivot table as a bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6))

sns.despine(top=True, right=True)
# Adding title and labels
plt.title('Mutation Counts by Country')
plt.xlabel('Country')
plt.ylabel('Count of Mutations')

# Show the plot
plt.tight_layout()  # Ensures everything fits nicely
plt.show()
```

```python
# Create the heatmap
plt.figure(figsize=(12, 8))  # Adjust the figure size as needed
sns.heatmap(pivot_table, annot=True, cmap='Blues', fmt='g', cbar=True)

# Set labels and title
plt.title("Heatmap of K13 Validated Mutations", fontsize=16)
plt.xlabel("K13 Validated Mutations", fontsize=12)
plt.ylabel("Country and PMID", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
