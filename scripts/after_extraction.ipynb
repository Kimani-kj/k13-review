```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Find the earliest year for each country
min_years = first_reports.groupby('country')['year_pub'].min().reset_index()
min_years.columns = ['country', 'earliest_year']

# Step 2: Filter all records from that year per country
first_per_country_multi = first_reports.merge(min_years, on='country')
first_per_country_multi = first_per_country_multi[
    first_per_country_multi['year_pub'] == first_per_country_multi['earliest_year']
]

# Step 3: Combine mutations per country-year into a single row
grouped = first_per_country_multi.groupby(['country', 'year_pub', 'Longitude', 'Latitude']).agg({
    'k13_validated_mutations': lambda x: ', '.join(sorted(x.unique()))
}).reset_index()

# Step 4: Create GeoDataFrame
gdf = gpd.GeoDataFrame(
    grouped,
    geometry=gpd.points_from_xy(grouped.Longitude, grouped.Latitude),
    crs="EPSG:4326"
)

# Step 5: Load base world map
world = gpd.read_file('../ne_110m_admin_0_map_units/ne_110m_admin_0_map_units.shp')

# Step 6: Plot
fig, ax = plt.subplots(figsize=(15, 8))
world.plot(ax=ax, color='lightblue', edgecolor='white')

# Plot a single point per country
gdf.plot(ax=ax, color='blue', markersize=100)

# Step 7: Annotate with combined mutation + year + country
for idx, row in gdf.iterrows():
    annotation_text = f"{row['k13_validated_mutations']} \n({row['year_pub']})\n{row['country']}"
    ax.annotate(annotation_text,
                xy=(row.geometry.x, row.geometry.y),
                xytext=(3, 3),
                textcoords="offset points",
                fontsize=10)

# Optional zoom (adjust to your region)
ax.set_xlim(12, 55)
ax.set_ylim(-12, 23)

# Title and save
plt.title('Earliest Reported K13 Validated Mutations by Country')
plt.tight_layout()
plt.savefig("../results/first_report_combined_mutations.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge
import numpy as np

# ----------------------------
# SETUP MUTATION COLORS
# ----------------------------
mutation_colors = {
    'A675V': 'red',
    'R561H': 'blue',
    'R622I': 'green',
    'C469Y': 'purple',
    'P574L': 'orange',
    'R539T': 'palegreen',
    'Y493H': 'lightgrey',
    'M476I': 'black',
    'I543T': 'yellow'    
}

valid_mutations = list(mutation_colors.keys()) 

# ----------------------------
# CLEAN DATA
# ----------------------------
df_clean = df[df['k13_validated_mutations'].isin(valid_mutations)].dropna(subset=['Longitude', 'Latitude'])

# Group unique mutations per country
mutations_per_country = (
    df_clean.groupby('country')['k13_validated_mutations']
    .apply(lambda x: sorted(set(x)))  # unique mutations only
    .reset_index()
)

# Get first coordinate per country (based on earliest year_pub)
coords = (
    df_clean.sort_values(by='year_pub')
            .drop_duplicates(subset='country', keep='first')
            [['country', 'Latitude', 'Longitude']]
)

# Merge mutations + coordinates
plot_data = mutations_per_country.merge(coords, on='country', how='inner')

# ----------------------------
# LOAD WORLD MAP
# ----------------------------
world = gpd.read_file('../ne_110m_admin_0_map_units/ne_110m_admin_0_map_units.shp')

# ----------------------------
# PIE DRAWING FUNCTION
# ----------------------------
def draw_manual_pie(ax, x, y, mutations, radius=1):
    num = len(mutations)
    if num == 0:
        return
    angle_per = 360 / num
    start_angle = 0
    for mut in mutations:
        wedge = Wedge(
            center=(x, y),
            r=radius,
            theta1=start_angle,
            theta2=start_angle + angle_per,
            facecolor=mutation_colors.get(mut, 'gray'),
            edgecolor='black',
            lw=0.5
        )
        ax.add_patch(wedge)
        start_angle += angle_per

# ----------------------------
# DRAW PIES
# ----------------------------
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='lightblue', edgecolor='white')

# Draw the pies based on the data
for _, row in plot_data.iterrows():
    draw_manual_pie(ax, row['Longitude'], row['Latitude'], row['k13_validated_mutations'], radius=1)

# ----------------------------
# FINAL TOUCHES
# ----------------------------
# Prevent autoscaling of the plot
ax.set_aspect('equal', adjustable='box')
ax.autoscale(enable=False)  # Disable autoscale

# Zoom to specific region (Africa)
ax.set_xlim(12, 55)  # Longitude range (from West to East)
ax.set_ylim(-12, 23)  # Latitude (South-North)

# Add legend for mutations
legend_handles = [
    plt.Line2D([0], [0], marker='o', color='w', label=mut, markerfacecolor=color, markersize=10)
    for mut, color in mutation_colors.items()
]
ax.legend(handles=legend_handles, title='K13 Mutations', loc='lower left', bbox_to_anchor=(1, 0))

# Annotate with country names
for _, row in plot_data.iterrows():
    ax.annotate(row['country'],
                xy=(row['Longitude'], row['Latitude']),
                xytext=(10, 10),  # Offset text slightly to avoid overlap
                textcoords="offset points",
                fontsize=11,  # Adjust font size
                color='black')

# Title and final layout
plt.title('K13 Validated Mutations per Country ', fontsize=14)
plt.tight_layout()
plt.savefig("../results/all_validated.png", dpi=300, bbox_inches='tight')
plt.show()
```
